```python
## assume daily returns for the S&P 500 are normally distributed with mean and standard deviation based on historical data. We will also assume the investor starts with a fixed initial investment
# Define simulation parameters
num_simulations = 10000
years = 2
days_per_year = 252  # Approximate number of trading days in a year
days_out_of_market = 10

# Historical daily mean and standard deviation of S&P 500 returns from the daily returns dataset
annual_mean_return = 0.000426186 #  average daily return
annual_std_deviation = 0.011475  # average daily standard deviation

# Initial investment amount
initial_investment = 1000  # Replace with your desired initial investment

# Simulate the investor's portfolio
terminal_wealths = []

for _ in range(num_simulations):
    terminal_wealth = initial_investment
    
    for _ in range(years):
        annual_returns = np.random.normal(annual_mean_return, annual_std_deviation, days_per_year)
        days_out = np.random.choice(days_per_year, days_out_of_market, replace=False)
        
        for day in range(days_per_year):
            if day not in days_out:
                terminal_wealth *= 1 + annual_returns[day]
    
    terminal_wealths.append(terminal_wealth)

# Calculate metrics
expected_terminal_wealth = np.mean(terminal_wealths)
median_terminal_wealth = np.median(terminal_wealths)

# Calculate the probability of beating the buy-and-hold strategy
buy_and_hold_wealth = initial_investment * (1 + annual_mean_return) ** years
probability_beating_buy_and_hold = np.mean(np.array(terminal_wealths) > buy_and_hold_wealth)

# Print results
print(f"Expected Terminal Wealth: ${expected_terminal_wealth:.2f}")
print(f"Median Terminal Wealth: ${median_terminal_wealth:.2f}")
print(f"Probability of Beating Buy-and-Hold: {probability_beating_buy_and_hold * 100:.2f}%")
```
